```python
# Install python libray that we need to execute this practice 

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from fake_useragent import UserAgent
```

```python
## address we want to access 
resource = requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en&vq=ko')

## convert resource to TEXT format
html = resource.text

# ua = UserAgent(verify_ssl=False)
# headers = {'User-Agent' : ua.random}
# html = requests.get(url, headers=headers)
```

```python
## convert html.text to bs4 format 
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
# <h3><b>First-Page </b></h3>

# collect data (‘name’, ’h5-index’, ‘h5-median’) of 100 Journals
```

```python
p_info = soup.find_all('tr')
p_info
```

```python
scholar_metrics = []
for p in p_info:
    rank = p.find('td',{'class':'gsc_mvt_p'})
    title = p.find('td',{'class':'gsc_mvt_t'})
    h5_indexMedian = p.find_all('td',{'class':'gsc_mvt_n'})
    
    if title!=None:
        scholar_metrics.append([rank.text,title.text,h5_indexMedian[0].text,h5_indexMedian[1].text])
```

```python
scholar_metrics
```

```python
scholar_metrics = pd.DataFrame(scholar_metrics)
```

```python
# <h3><b>Second-Page </b></h3>

#   collect data (‘title’, author, ‘cited by’) on each journal
```

```python
href_list = []
for p in p_info:
    h5_index = p.find('td',{'class':'gsc_mvt_n'})  
    if h5_index!=None:
        href = h5_index.find('a')['href']
        adr = 'https://scholar.google.com/' + href
        href_list.append(adr)
print(href_list)
```

```python
new=[]
for s in href_list:
    response = requests.get(s)
    time.sleep(20)
    if response.status_code ==200:
        second_soup = BeautifulSoup(response.text, 'html.parser')
        s_info = second_soup.select('#gsc_mpat_table>tr')
        for ss in s_info[1:]:
            # findall로 받아오면 resultset형태로 받아와서 .text라는 함수를 쓸수없습니다. 
            # 그래서 find_all 쓰고 title[0].text로 받아오던지 해야합니다.
            # 저는 그냥 a태그로 받아왔습니다. 
            title = ss.find('a',{'class':'gsc_mp_anchor_lrge'})
            author = ss.find('div',{'class':'gs_gray'})
            cited_by = ss.find('a',{'class':'gs_ibl gsc_mp_anchor'})
            new.append(ss)
```
